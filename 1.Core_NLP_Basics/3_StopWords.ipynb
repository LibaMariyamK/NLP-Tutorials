# ✅ What are **Stopwords**?

**Stopwords = common words that do not add meaningful information** in text analysis.

Examples in English:
`a, an, the, is, am, are, was, and, but, on, in`

* They appear very frequently
* Usually **ignored in NLP tasks** like:

  * Text classification
  * Search engines
  * Keyword extraction

**Why remove them?**

* Reduces noise
* Speeds up processing
* Focuses on **meaningful words**

# ✅ Stopword Removal in NLTK

### Step 1: Import and download stopwords

```python
import nltk
nltk.download('stopwords')   # Download stopwords corpus
```

```python

from nltk.corpus import stopwords
```

### Step 2: Get English stopwords

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

### Step 3: Remove stopwords from a sentence

```python
sentence = "This is a simple example showing the removal of stopwords."
sentence = sentence.lower()
print(sentence)
```

```python
from nltk.tokenize import word_tokenize

words = word_tokenize(sentence)

print(words)
```

```python
words = sentence.split()

filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

✅ Only meaningful words are left.

# ✅ Stopword Removal in spaCy

spaCy has **built-in stopwords**, which makes it easy:

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("This is a simple example showing the removal of stopwords.")

filtered_words = [token.text for token in doc if not token.is_stop]
print(filtered_words)
```

* `token.is_stop` automatically identifies stopwords
* Works with spaCy’s tokenizer

```python
# Remove stopwords and lemmatize at the same time
filtered_lemmas = [token.lemma_ for token in doc if not token.is_stop]

print(filtered_lemmas)
```


## ✅ Key Notes

1. Stopwords are **language-dependent**.

   * English has many pre-defined stopwords in NLTK and spaCy
   * For other languages, you may need a **custom list** or a library like `stopwordsiso`.

2. Stopwords are **optional**:

   * Some NLP tasks like sentiment analysis might keep them.
   * In some deep learning pipelines, models can learn to ignore them automatically.

3. NLTK vs spaCy:
   | Feature | NLTK | spaCy |
   |---------|------|-------|
   | Stopword list | Predefined | Predefined |
   | Check word | `word in stop_words` | `token.is_stop` |
   | Ease of use | Moderate | Very easy |

## ✅ Custom Stopwords

You can always **add your own stopwords**:

```python
custom_stops = set(stop_words)
custom_stops.update(["example", "showing"])  # add extra words to remove

filtered_words = [w for w in words if w.lower() not in custom_stops]
print(filtered_words)
```
